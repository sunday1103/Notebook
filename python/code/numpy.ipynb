# 基本矩阵和矩阵生成函数

```python
'基本的一维二维三维矩阵'
import numpy as np
a = np.array([0, 1, 2, 3])
print(type(a))
print(repr(a))
print(str(a))
a

b = np.array([[0,1,2],[2,3,4]])
print(repr(b))
print(b.ndim)
print(b.shape)
print(len(b))
```

```python
'生成矩阵函数'
a = np.arange(10)
print(a)
b = np.arange(1,11,2) # 不包括11
print(b)

c = np.linspace(0,1,6)
print(c)
d = np.linspace(0,1,5, endpoint = False)
print(d)

a = np.ones((3,3))
print(a)
# a1 = np.ones(3,3)  data type not understood
aa = np.ones_like(a)
print(aa)

b = np.zeros((2,2))
print(b)
bb = np.zeros_like(b)
print(bb)

c = np.eye(3)
print(c)

d = np.diag(np.array([1,2,3,4]))
print(d)
```

```python
'随机数'
a = np.random.rand(4)
print(a)

b = np.random.randn(4)
print(b)
```

```python
'基本数据类型'
a = np.array([1,2,3])
print(a.dtype)

b = np.array([1.0, 2, 2])
print(b.dtype)

c = np.array([1, 2, 3], dtype = float)
c.dtype
```

# 切片和索引

```python
'索引和切片 indexing and slicing'
'一维'
a = np.arange(10)
print(a)
print(a[:4]) # not include 4
b = a[::2]
print(b)
b[1] = 10
print(b)
print(a)
```

```python
'多维'
a1 = [0,1,2,3]
a2 = [2,3,4,5]
a3 = [3,4,5,5]
a = np.array([a1,a2,a3])
print(a)
print(a[1,1:3])
b = a[1:3,2:4]
print(b)
```

```python
'''通过切片和索引获取的矩阵是引用
copy 或者 view
'''
a = np.arange(10)
b = a[::2]
print(np.may_share_memory(a,b))
b[0] = 10
print(a)
'a 和 b共享存储'

a = np.arange(10)
c = a[::2].copy()
print(c)
c[0] = 10
print(a)
```

# 高级切片特性 fancy indexing

```python
'用boolean变量作为掩码 marks'
a = np.array([1,2,3,4,5,6])
m = np.array([True, False, True, True, False, False])  # 注意是np的array作为掩码
print(a[m])

'用掩码赋值'
print(a)
a[m] = 8
print(a)
print()

'用Python list作为 矩阵索引'
a = np.array([1,2,3,4,5,6])
b = [1,2,2,0,1,1]
print(a[b])

'赋值'
a[b] = 100
print(a)
```

# 矩阵上的基本计算

```python
a = np.array([1,2,3,4])
print(a + 1)
print(2**a)

b = np.ones(4) + 1
print(a - b)
print(a * b)
```

```python
c = np.array([[0,1],[2,3]])
print(c * c) # 不是矩阵乘法，而是普通的按照位置乘法
print(c.dot(c)) # 矩阵乘法
```

```python
a = np.array([1,2,3,4])
b = np.array([4,2,2,4])

print(a==b)
print(a > b)
print(np.array_equal(a,b))
print(np.array_equal(a,a))


a = np.array([1,1,0,0])
b = np.array([1,0,1,0])
print(np.logical_or(a,b))
print(np.logical_and(a,b))

a = np.arange(10)
print(np.sin(a))
print(np.log(a))
print(np.exp(a))

```
